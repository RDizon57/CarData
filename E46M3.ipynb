```python
import tkinter as tk
from tkinter import *
import pandas as pd
import time
import matplotlib
import matplotlib.pyplot as plt
matplotlib.use("TkAgg")
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure
print('all packages have arrived!')

# setup the main window
m = Tk()
t = Label(m, text='E46 M3 Data')
t.grid()

# name all the entry info of the listing
Label(m, text='Year').grid(row=1) 
Label(m, text='Trans').grid(row=2) 
Label(m, text='Mileage').grid(row=3)
Label(m, text='Color').grid(row=4)
Label(m, text='Body Style').grid(row=5)
Label(m, text='Location').grid(row=6)
Label(m, text='Price').grid(row=7)

# use Entry to input single line text entries for each info line
e1 = tk.Entry(m) 
e2 = tk.Entry(m) 
e3 = tk.Entry(m) 
e4 = tk.Entry(m) 
e5 = tk.Entry(m) 
e6 = tk.Entry(m) 
e7 = tk.Entry(m) 
e1.grid(row=1, column=1) 
e2.grid(row=2, column=1)
e3.grid(row=3, column=1) 
e4.grid(row=4, column=1) 
e5.grid(row=5, column=1) 
e6.grid(row=6, column=1) 
e7.grid(row=7, column=1)

# setup the car dataframe   
df = pd.read_csv('C:\\Users\Ricky\Documents\CarData\E46M3Data.csv') # ENTER YOUR PATHWAY HERE to read the csv
print(df)

# count the number of listings already in the csv so we can add to it
listing = 0
for row in df.index:
    listing = listing + 1


# define the function:
def cardata():
    
    # setup the parameters of the listing
    df.at[listing, 'Year'] = e1.get()
    df.at[listing, 'Trans'] = e2.get()
    df.at[listing, 'Mileage'] = e3.get()
    df.at[listing, 'Color'] = e4.get()
    df.at[listing, 'Body Style'] = e5.get()
    df.at[listing, 'Location'] = e6.get()
    df.at[listing, 'Price'] = e7.get()
    df['Mileage'] = df['Mileage'].astype('int') # define mileage, year, and price as ints while keeping the rest as objects
    df['Price'] = df['Price'].astype('int')
    df['Year'] = df['Year'].astype('int')
    
    # create a new window for your graphs:
    graphs = tk.Toplevel(m) # create a toplevel window separate from the main window
    
    # setup figure 1
    figure1 = plt.Figure(figsize=(5,5), dpi=100)
    ax1 = figure1.add_subplot(1,1,1)
    
    # display statistics about the specific listing compared to the rest of the dataframe:
    MeanPrice = df['Price'].mean()
    MeanMiles = df['Mileage'].mean()
    stdPrice =(df.at[listing, 'Price'] - MeanPrice)/df['Price'].std() # standard deviation away from the mean
    stdMiles =(df.at[listing, 'Mileage'] - MeanMiles)/df['Mileage'].std() 
    ax1.annotate('Mean Price [$] =', xy=(10, 350), xycoords = 'figure points') 
    ax1.annotate(MeanPrice, xy=(95, 350), xycoords = 'figure points')
    ax1.annotate('Mean Mileage [mi] =', xy=(10, 340), xycoords = 'figure points') 
    ax1.annotate(MeanMiles, xy=(115, 340), xycoords = 'figure points')
    ax1.annotate('Price STD =', xy=(185, 350), xycoords = 'figure points') 
    ax1.annotate(stdPrice, xy=(245, 350), xycoords = 'figure points')
    ax1.annotate('Mileage STD =', xy=(185, 340), xycoords = 'figure points') 
    ax1.annotate(stdMiles, xy=(260, 340), xycoords = 'figure points')

    # create a scatter plot for Mileage vs Price:
    ax1.scatter(df['Mileage']*10**-3, df['Price'], color = 'b') 
    ax1.scatter(df.at[listing, 'Mileage']*10**-3, df.at[listing, 'Price'], color = 'r') # set the latest data point to red
    scatter1 = FigureCanvasTkAgg(figure1, graphs) 
    scatter1.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    ax1.legend(['Price [$]']) 
    ax1.set_xlabel('Mileage [mi x 1000]')
    ax1.set_title('Mileage vs Price')
    ax1.set_xlim((50,200)) # limit the x axis
    ax1.set_ylim((0,30000))
    
    # setup figure 2
    figure2 = plt.Figure(figsize=(5,5), dpi=100)
    ax2 = figure2.add_subplot(1,1,1)

    # create a scatter plot for Year vs Price:
    ax2.scatter(df['Year'],df['Price'], color = 'b')
    ax2.scatter(df.at[listing, 'Year'], df.at[listing, 'Price'], color = 'r') # set the latest data point to red
    scatter2 = FigureCanvasTkAgg(figure2, graphs) 
    scatter2.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    ax2.legend(['Price [$]'])
    ax2.set_xlabel('Year')
    ax2.set_title('Year vs Price')
    ax2.set_xlim((2000,2006)) # limit the x axis
    ax2.set_ylim((0,30000)) 
    
    plt.rcParams.update({'font.size': 10}) # edit the font size to size 10
    
    df.to_csv('C:\\Users\Ricky\Documents\CarData\E46M3Data.csv',index=False) # ENTER YOUR PATHWAY HERE TO overwrite the csv with new data
       
def Years():
    graphs3 = tk.Toplevel(m) # create a toplevel window separate from the main window
    figure3 = plt.Figure(figsize=(5,5), dpi=100)
    ax3 = figure3.add_subplot(1,1,1)
    ax3.hist(df['Year'])
    hist3 = FigureCanvasTkAgg(figure3, graphs3) 
    hist3.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    ax3.legend(['Frequency']) 
    ax3.set_xlabel('Year')
    ax3.set_title('Years')
    ax3.set_xlim((2000,2006))
    
def Trans():
    graphs4 = tk.Toplevel(m) # create a toplevel window separate from the main window
    figure4 = plt.Figure(figsize=(5,5), dpi=100)
    
    ax4 = figure4.add_subplot(1,1,1)
    ax4.hist(df['Trans'])
    hist4 = FigureCanvasTkAgg(figure4, graphs4) 
    hist4.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    ax4.legend(['Frequency']) 
    ax4.set_xlabel('Trans')
    ax4.set_title('Transmissions')
    
    figure10 = plt.Figure(figsize=(5,5), dpi=100)
    ax10 = figure10.add_subplot(1,1,1)
    
    ax10.scatter(df['Trans'], df['Price'], color = 'b')
    scatter10 = FigureCanvasTkAgg(figure10, graphs4) 
    scatter10.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    ax10.legend(['Price [$]'])
    ax10.set_xlabel('Trans')
    ax10.set_title('Trans vs Price')
    ax10.set_ylim((0,30000))
    
def Color():
    graphs5 = tk.Toplevel(m) # create a toplevel window separate from the main window
    figure5 = plt.Figure(figsize=(5,5), dpi=100)
    ax5 = figure5.add_subplot(1,1,1)
    ax5.hist(df['Color'])
    hist5 = FigureCanvasTkAgg(figure5, graphs5) 
    hist5.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    ax5.legend(['Frequency']) 
    ax5.set_xlabel('Color')
    ax5.set_title('Colors')
    
    figure11 = plt.Figure(figsize=(5,5), dpi=100)
    ax11 = figure11.add_subplot(1,1,1)
    
    ax11.scatter(df['Color'], df['Price'], color = 'b')
    scatter11 = FigureCanvasTkAgg(figure11, graphs5) 
    scatter11.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    ax11.legend(['Price [$]'])
    ax11.set_xlabel('Color')
    ax11.set_title('Color vs Price')
    ax11.set_ylim((0,30000))
    
def Bodystyle():
    graphs6 = tk.Toplevel(m) # create a toplevel window separate from the main window
    figure6 = plt.Figure(figsize=(5,5), dpi=100)
    ax6 = figure6.add_subplot(1,1,1)
    ax6.hist(df['Body Style'])
    hist6 = FigureCanvasTkAgg(figure6, graphs6) 
    hist6.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    ax6.legend(['Frequency']) 
    ax6.set_xlabel('Body Style')
    ax6.set_title('Body Styles')
    
    figure12 = plt.Figure(figsize=(5,5), dpi=100)
    ax12 = figure12.add_subplot(1,1,1)
    
    ax12.scatter(df['Body Style'], df['Price'], color = 'b')
    scatter12 = FigureCanvasTkAgg(figure12, graphs6) 
    scatter12.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    ax12.legend(['Price [$]'])
    ax12.set_xlabel('Body Style')
    ax12.set_title('Body Style vs Price')
    ax12.set_ylim((0,30000))
    
def Location():
    graphs7 = tk.Toplevel(m) # create a toplevel window separate from the main window
    figure7 = plt.Figure(figsize=(5,5), dpi=100)
    ax7 = figure7.add_subplot(1,1,1)
    ax7.hist(df['Location'])
    hist7 = FigureCanvasTkAgg(figure7, graphs7) 
    hist7.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    ax7.legend(['Frequency']) 
    ax7.set_xlabel('Location')
    ax7.set_title('Locations')
    
    figure13 = plt.Figure(figsize=(5,5), dpi=100)
    ax13 = figure13.add_subplot(1,1,1)
    
    ax13.scatter(df['Location'], df['Price'], color = 'b')
    scatter13 = FigureCanvasTkAgg(figure13, graphs7) 
    scatter13.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    ax13.legend(['Price [$]'])
    ax13.set_xlabel('Location')
    ax13.set_title('Location vs Price')
    ax13.set_ylim((0,30000))
    
def Current():
    graphs8 = tk.Toplevel(m) # create a toplevel window separate from the main window
    figure8 = plt.Figure(figsize=(5,5), dpi=100)
    ax8 = figure8.add_subplot(1,1,1)
    
    MeanPrice = df['Price'].mean()
    MeanMiles = df['Mileage'].mean()
    ax8.annotate('Mean Price [$] =', xy=(10, 350), xycoords = 'figure points') 
    ax8.annotate(MeanPrice, xy=(95, 350), xycoords = 'figure points')
    ax8.annotate('Mean Mileage [mi] =', xy=(10, 340), xycoords = 'figure points') 
    ax8.annotate(MeanMiles, xy=(115, 340), xycoords = 'figure points')
    
    ax8.scatter(df['Mileage']*10**-3, df['Price'], color = 'b') 
    scatter8 = FigureCanvasTkAgg(figure8, graphs8) 
    scatter8.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    ax8.legend(['Price [$]']) 
    ax8.set_xlabel('Mileage [mi x 1000]')
    ax8.set_title('Mileage vs Price')
    ax8.set_xlim((50,200)) # limit the x axis
    ax8.set_ylim((0,30000))
    
    # setup figure 9
    figure9 = plt.Figure(figsize=(5,5), dpi=100)
    ax9 = figure9.add_subplot(1,1,1)

    # create a scatter plot for Year vs Price:
    ax9.scatter(df['Year'], df['Price'], color = 'b')
    scatter9 = FigureCanvasTkAgg(figure9, graphs8) 
    scatter9.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    ax9.legend(['Price [$]'])
    ax9.set_xlabel('Year')
    ax9.set_title('Year vs Price')
    ax9.set_xlim((2000,2006)) # limit the x axis
    ax9.set_ylim((0,30000))
    
    plt.rcParams.update({'font.size': 10}) # edit the font size to size 10
       
    
# setup a button to submit the listing info
b = tk.Button(m, text='Enter', width=25, command=cardata) # command is used to call a function
b.grid(row=8, column=1) 
b = tk.Button(m, text='Current Data', width=25, command=Current) 
b.grid(row=9, column=1)
b = tk.Button(m, text='Years', width=25, command=Years) 
b.grid(row=10, column=1)
b = tk.Button(m, text='Trans', width=25, command=Trans) 
b.grid(row=11, column=1)
b = tk.Button(m, text='Colors', width=25, command=Color) 
b.grid(row=12, column=1)
b = tk.Button(m, text='Body Style', width=25, command=Bodystyle) 
b.grid(row=13, column=1)
b = tk.Button(m, text='Locations', width=25, command=Location) 
b.grid(row=14, column=1)
    
m.mainloop() 
```
